# Computing Trajectories: 2nd-Order Systems (Acceleration Field)
In this notebook, we will look at some basic numerical methods for calculating trajectories of particles determined by an acceleration field, which in physics is derived from a force law. We will implement the methods manually by ourselves. 

Some terminology. 
- <b>first-order:</b> up to first-order derivatives only.
- <b>second-order:</b> up to second-order derivatives only.
- <b>ordinary:</b> total derivatives only.
- <b>partial:</b> partial derivatives included.
- <b>linear:</b> unknown function(s) $y(t)$ and its derivatives appear only once in each term in the form $a(t)y^{(n)}(t)$; no products between $y(t)$ and its derivatives (e.g. $y(t)^{2}$ or $y(t)y'(t)$); $y(t)$ is not in the argument of some nonlinear function (e.g. $e^{y(t)}$). 
- <b>non-linear:</b> not linear. 
- <b>homogeneous:</b> there is no term that does not involve the unknown function $y(t)$ (we call such a term the "constant term" even though it may depend on $t$).
- <b>non-homogeneous:</b> there is a nonzero so-called "constant term."

The systems we will examine here can be formulated in at least two ways. 
- <b>First way:</b> The systems are obeying the second-order ODE 
\begin{align*}
    \ddot{\vec{x}}(t) = \vec{a}(\vec{x}(t))
\end{align*}
with initial conditions $\vec{x}(0) = \vec{x}_{0}$ and $\dot{\vec{x}}(0) = \vec{v}_{0}$.
- <b>Second way:</b> The systems are obeying the system of two coupled first-order ODEs
\begin{align*}
    \dot{\vec{x}}(t) &= \vec{v}(t), \\
    \dot{\vec{v}}(t) &= \vec{a}(\vec{x}(t))
\end{align*}
with initial conditions $\vec{x}(0) = \vec{x}_{0}$ and $\vec{v}(0) = \vec{v}_{0}$. 

We will use whichever formulation is more convenient at hand. The example we will use will be the case of particles orbiting a central massive gravitating body:
$$ \vec{a}(\vec{x}) = -\frac{GM}{x^{3}}\vec{x} $$
where $G, M$ are some constants. 

In this case, we want to see how well the numerical methods create orbital trajectories. For this we will write down the analytical solution and plot it first, so that the calculated trajectories are compared to the analytical one for reference. 

<b>Elliptical Orbits.</b>

Consider an ellipse centered at $(0, 0)$ with vertices $(\pm a, 0)$ and co-vertices $(0, \pm b)$. Its foci are located at $(\pm c, 0)$ where $c^{2} = a^{2} - b^{2}$. The eccentricity is 
$$ e = \frac{c}{a} = \sqrt{1 - \frac{b^{2}}{a^{2}}}. $$ 

In our example, we will be interested in an elliptical orbit with $a = 3$ and $b = 2$. If the ellipse is relocated so that one of its foci is placed at the origin and the major axis is aligned with the $x$-axis, its equation is
$$ \frac{(x-c)^{2}}{9} + \frac{y^{2}}{4} = 1 $$
with $c = \sqrt{5} \approx 2.24$. This has $e = \sqrt{5}/3 \approx 0.75$.

```python
import numpy as np
import matplotlib.pyplot as plt

# plot the ellipse
theta = np.linspace(0, 2*np.pi, 100)

plt.grid(color='lightgray', linestyle='--', zorder=0)
plt.xlim(-6, 6), plt.ylim(-6, 6)
plt.xticks(np.arange(-6, 7, 1)), plt.yticks(np.arange(-6, 7, 1))
plt.xlabel("x"), plt.ylabel("y")
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

plt.plot(3.00*np.cos(theta)+2.24, 2.00*np.sin(theta), color='red', zorder=5)
plt.scatter([0], [0], color='k', zorder=5) # large gravitating mass
plt.scatter([5.24, -0.76], [0, 0], color='red', zorder=5) # apogee and perigee
plt.show()
```

In our example, we will take $GM = 5$ so that our acceleration field is given as
$$ \vec{a}(\vec{x}) = -\frac{5}{x^{3}}\vec{x}, $$
and we will take the orbiting particle to have mass $m = 1$. In order for the particle to take the desired elliptical orbit, we take the initial conditions to be
$$ \vec{x}(0) = (5.24, 0) \quad\text{ and }\quad \vec{v}(0) = (0, 0.49) $$
where the initial velocity was chosen so that the trajectory would match the given ellipse by using the vis-viva equation
$$ v^{2} = GM\left( \frac{2}{r} - \frac{1}{a} \right) $$
with $GM = 5$, $r = a+c$, $a = 3$, and $c = \sqrt{5}$. 


## Method 1: Euler

Given our system
\begin{align*}
    \dot{\vec{x}}(t) &= \vec{v}(t), \\
    \dot{\vec{v}}(t) &= \vec{a}(\vec{x}(t)),
\end{align*}
we descritize this with a finite nonzero step $\Delta t$ to obtain
\begin{gather}
    \boxed{\vec{x}_{n+1} = \vec{x}_{n} + \vec{v}_{n}\Delta t, \\
    \vec{v}_{n+1} = \vec{v}_{n} + \vec{a}(\vec{x}_{n})\Delta t}
\end{gather}
with initial conditions $\vec{x}_{0} = (5.24, 0)$ and $\vec{v}_{0} = (0, 0.49)$.

The local error here is $O(\Delta t^{2})$. As we will see, the calculated trajectory deviates away from the ellipse noticably even for $\Delta t\sim 0.01$, and the deviation gets arbitrarily large in time. This is a major issue.  Physically, the energy of the orbiting particle is not conserved under this numerical method, so Euler's method is not energy preserving. Other methods are able to overcome this issue to varying degrees, and we will see how well they perform in this regard. 

```python
# Euler
# step and initial conditions
totaltime = 10
dt = 0.1
x, y = np.array([5.24]), np.array([0.00])
vx, vy = np.array([0.00]), np.array([0.49])

# update rule
def accel_law(x, y):
    b = 5/np.power(x**2 + y**2, 1.5)
    return -b*x, -b*y

def Euler_update(x, y, vx, vy, dt):
    ax, ay = accel_law(x[-1], y[-1])
    newx, newy = x[-1] + vx[-1]*dt, y[-1] + vy[-1]*dt
    newvx, newvy = vx[-1] + ax*dt, vy[-1] + ay*dt
    return np.append(x, newx), np.append(y, newy), np.append(vx, newvx), np.append(vy, newvy)

# create tajectory; should take `totaltime` seconds
for i in range(0, round(totaltime/dt)):
    x, y, vx, vy = Euler_update(x, y, vx, vy, dt)

# create chart with ellipse
plt.grid(color='lightgray', linestyle='--', zorder=0)
plt.xlim(-6, 6), plt.ylim(-6, 6)
plt.xticks(np.arange(-6, 7, 1)), plt.yticks(np.arange(-6, 7, 1))
plt.xlabel("x"), plt.ylabel("y")
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(3.00*np.cos(theta)+2.24, 2.00*np.sin(theta), color='red', zorder=5)
plt.scatter([0], [0], color='k', zorder=5) # large gravitating mass

# plot the trajectory
plt.plot(x, y, color='green', label="step = {}".format(dt), zorder=10)
plt.scatter([x[0]], [y[0]], color='green', zorder=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Euler method")
plt.show()
```

## Method 2: Midpoint AKA Runge-Kutta 2-nd Order (RK2)

Let us see how the midpoint method improves upon Euler's method in this situation, and whether it solves the energy drift issue.

In Notebook 1, we used it for a particle moving in a velocity field. In that scenario, we took the midpoint of the usual Euler's method update $\vec{x}_{n}\rightarrow x_{n} + \vec{v}(\vec{x}_{n})\Delta t$ to obtain $\text{midpt} = \vec{x}_{n} + \vec{v}_{n}\tfrac{\Delta t}{2}$ ($\vec{v}_{n} = \vec{v}(\vec{x}_{n})$), looked at the velocity at that point to get $\vec{v}(\vec{x}_{n} + \vec{v}_{n}\tfrac{\Delta t}{2})$, and then made the update $\vec{x}_{n}\rightarrow \vec{x}_{n} + \vec{v}(\vec{x}_{n} + \vec{v}_{n}\tfrac{\Delta t}{2})$. 

How can the midpoint method apply in the case where we have an acceleration field, and both position and velocity are variables to be solved for or numerically computed for the trajectory? To answer this, we will proceed by analogy to what we did in Notebook 1. Let us start with the Euler's method update
\begin{align*}
    \vec{x}_{n} &\rightarrow \vec{x}_{n} + \vec{v}_{n}\Delta t, \\
    \vec{v}_{n} &\rightarrow \vec{v}_{n} + \vec{a}(\vec{x}_{n})\Delta t.
\end{align*}
Then we take the midpoint of the updates to both variables, so we take
\begin{align*}
    \vec{x}_{n+1/2} &= \vec{x}_{n} + \vec{v}_{n}\tfrac{\Delta t}{2}, \\
    \vec{v}_{n+1/2} &= \vec{v}_{n} + \vec{a}(\vec{x}_{n})\tfrac{\Delta t}{2}.
\end{align*}
These are the so-called **intermediate midpoints** that we will use. We modify the usual Euler's method update rule by plugging these new intermediate variables as follows:
\begin{align*}
    \vec{x}_{n+1} &= \vec{x}_{n} + \vec{v}_{n+1/2}\Delta t, \\
    \vec{v}_{n+1} &= \vec{v}_{n} + \vec{a}(\vec{x}_{n+1/2})\Delta t.
\end{align*}

So let's compile what we have together. We take
\begin{gather}
    \boxed{\vec{x}_{n+1} = \vec{x}_{n} + \vec{v}_{n+1/2}\Delta t, \\
    \vec{v}_{n+1} = \vec{v}_{n} + \vec{a}(\vec{x}_{n+1/2})\Delta t}
\end{gather}
where
\begin{gather}
    \boxed{\vec{x}_{n+1/2} = \vec{x}_{n} + \vec{v}_{n}\tfrac{\Delta t}{2}, \\
    \vec{v}_{n+1/2} = \vec{v}_{n} + \vec{a}(\vec{x}_{n})\tfrac{\Delta t}{2}}
\end{gather}
and the initial conditions are $\vec{x}_{0} = (5.24, 0)$ and $\vec{v}_{0} = (0, 0.49)$. 

The local error here is $O(\Delta t^{3})$. Althugh this helps somewhat reduce the error without needing to decrease the step $\Delta t$, it does not solve the energy drift problem, as the reader can see below. 

(We will skip the Runge-Kutta 4-th Order method in this notebook, because it is too cumbersome and it won't provide us with much insight here. In any case, an understanding of RK2 in this and the previous notebook along with an understanding of RK4 from the previous notebook should be enough for the reader to figure out how RK4 can be implemented for this system if they are interested.)

```python
# midpoint AKA RK2
# step and initial conditions
totaltime = 10
dt = 0.1
x, y = np.array([5.24]), np.array([0.00])
vx, vy = np.array([0.00]), np.array([0.49])

# update rule
def accel_law(x, y):
    b = 5/np.power(x**2 + y**2, 1.5)
    return -b*x, -b*y

def midpoint_update(x, y, vx, vy, dt):
    ax, ay = accel_law(x[-1], y[-1])
    kx1, ky1 = x[-1] + vx[-1]*dt/2, y[-1] + vy[-1]*dt/2 # intermediate midpoints
    kvx1, kvy1 = vx[-1] + ax*dt/2, vy[-1] + ay*dt/2
    ax, ay = accel_law(kx1, ky1)
    newx, newy = x[-1] + kvx1*dt, y[-1] + kvy1*dt # new values
    newvx, newvy = vx[-1] + ax*dt, vy[-1] + ay*dt
    return np.append(x, newx), np.append(y, newy), np.append(vx, newvx), np.append(vy, newvy)

# create tajectory; should take `totaltime` seconds
for i in range(0, round(totaltime/dt)):
    x, y, vx, vy = midpoint_update(x, y, vx, vy, dt)

# create chart with ellipse
plt.grid(color='lightgray', linestyle='--', zorder=0)
plt.xlim(-6, 6), plt.ylim(-6, 6)
plt.xticks(np.arange(-6, 7, 1)), plt.yticks(np.arange(-6, 7, 1))
plt.xlabel("x"), plt.ylabel("y")
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(3.00*np.cos(theta)+2.24, 2.00*np.sin(theta), color='red', zorder=5)
plt.scatter([0], [0], color='k', zorder=5) # large gravitating mass

# plot the trajectory
plt.plot(x, y, color='green', label="step = {}".format(dt), zorder=10)
plt.scatter([x[0]], [y[0]], color='green', zorder=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Midpoint method")
plt.show()
```

## Method 3: Leapfrog

Here is another method to improve upon Euler's method and the midpoint method. Instead of looking $\vec{x}_{n}$ and $\vec{v}_{n}$ based off of the given point or an intermediate midpoint, we will update the $\vec{x}_{n}$'s by intermediate $\vec{v}_{n+1/2}$'s and update the intermediate $\vec{v}_{n+1/2}$'s by the $\vec{x}_{n}$. 

<img src="leapfrog_img.png" alt="Drawing" style="width: 600px;"/>

Given initial conditions $\vec{x}_{0}, \vec{v}_{0}$, we need to do a special initial update to velocity $\vec{v}_{0}\rightarrow\vec{v}_{1/2}$. With this in mind, we take
\begin{gather}
    \boxed{ \vec{x}_{n+1} = \vec{x}_{n} + \vec{v}_{n+1/2}\Delta t, \\
    \vec{v}_{n+1/2} = \vec{v}_{n-1/2} + \vec{a}(\vec{x}_{n})\Delta t }
\end{gather}
where
\begin{gather}
    \boxed{ \vec{v}_{1/2} = \vec{v}_{0} + \vec{a}(\vec{x}_{0})\tfrac{\Delta t}{2} }
\end{gather}
and the initial conditions are $\vec{x}_{0} = (5.24, 0)$ and $\vec{v}_{0} = (0, 0.49)$. 

The local error here is $O(\Delta t^{3})$. This method has the property that if we reversed the update rule, the equations for the backward update would take the same form as the equations for the forward update. Thus we say this method is **time-reversible**. For **periodic trajectories**, time-reversibility prevents long-term energy drift (the same applies to conservation of angular momentum and any other quantity that is supposed to be conserved). There can however be a periodic energy error. 

```python
# leapfrog
# step and initial conditions
totaltime = 10
dt = 0.1
x, y = np.array([5.24]), np.array([0.00])
vx, vy = np.array([0.00]), np.array([0.49])

# update rule
def accel_law(x, y):
    b = 5/np.power(x**2 + y**2, 1.5)
    return -b*x, -b*y

def leapfrog_update(x, y, vx, vy, dt):
    ax, ay = accel_law(x[-1], y[-1])
    if len(vx) == 1 and len(vy) == 1: 
        # initial update; returns vel[1]=v_{1/2} and pos[1]=x_{1}
        newvx, newvy = vx[0] + ax*dt/2, vy[0] + ay*dt/2
        newx, newy = x[0] + newvx*dt, y[0] + newvy*dt
        return np.append(x, newx), np.append(y, newy), np.append(vx, newvx), np.append(vy, newvy)
    elif len(vx) > 1 and len(vy) > 1:
        # usual update; in general vel[n]=v_{n-1/2} and pos[n]=x_{n}
        newvx, newvy = vx[-1] + ax*dt, vy[-1] + ay*dt
        newx, newy = x[-1] + newvx*dt, y[-1] + newvy*dt
        return np.append(x, newx), np.append(y, newy), np.append(vx, newvx), np.append(vy, newvy)
    else:
        raise ValueError("velocity arrays need to be checked")

# create tajectory; should take `totaltime` seconds
for i in range(0, round(totaltime/dt)):
    x, y, vx, vy = leapfrog_update(x, y, vx, vy, dt)

# create chart with ellipse
plt.grid(color='lightgray', linestyle='--', zorder=0)
plt.xlim(-6, 6), plt.ylim(-6, 6)
plt.xticks(np.arange(-6, 7, 1)), plt.yticks(np.arange(-6, 7, 1))
plt.xlabel("x"), plt.ylabel("y")
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(3.00*np.cos(theta)+2.24, 2.00*np.sin(theta), color='red', zorder=5)
plt.scatter([0], [0], color='k', zorder=5) # large gravitating mass

# plot the trajectory
plt.plot(x, y, color='green', label="step = {}".format(dt), zorder=10)
plt.scatter([x[0]], [y[0]], color='green', zorder=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Leapfrog method")
plt.show()
```

## Method 4: Semi-Implicit Euler

In this method, we update the velocity first, and then use the updated velocity to update the position. We have
\begin{gather}
    \boxed{ \vec{v}_{n+1} = \vec{v}_{n} + \vec{a}(\vec{x}_{n})\Delta t, \\
    \vec{x}_{n+1} = \vec{x}_{n} + \vec{v}_{n+1}\Delta t }
\end{gather}
with initial conditions $\vec{x}_{0} = (5.24, 0)$ and $\vec{v}_{0} = (0, 0.49)$.

The local error here is $O(\Delta t^{2})$. It is of the same order as Euler's method, but it improves upon Euler's method for the same given $\Delta t$. 

```python
# semi-implicit Euler
# step and initial conditions
totaltime = 10
dt = 0.1
x, y = np.array([5.24]), np.array([0.00])
vx, vy = np.array([0.00]), np.array([0.49])

# update rule
def accel_law(x, y):
    b = 5/np.power(x**2 + y**2, 1.5)
    return -b*x, -b*y

def semi_implicit_update(x, y, vx, vy, dt):
    ax, ay = accel_law(x[-1], y[-1])
    newvx, newvy = vx[-1] + ax*dt, vy[-1] + ay*dt
    newx, newy = x[-1] + newvx*dt, y[-1] + newvy*dt
    return np.append(x, newx), np.append(y, newy), np.append(vx, newvx), np.append(vy, newvy)

# create tajectory; should take `totaltime` seconds
for i in range(0, round(totaltime/dt)):
    x, y, vx, vy = semi_implicit_update(x, y, vx, vy, dt)

# create chart with ellipse
plt.grid(color='lightgray', linestyle='--', zorder=0)
plt.xlim(-6, 6), plt.ylim(-6, 6)
plt.xticks(np.arange(-6, 7, 1)), plt.yticks(np.arange(-6, 7, 1))
plt.xlabel("x"), plt.ylabel("y")
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(3.00*np.cos(theta)+2.24, 2.00*np.sin(theta), color='red', zorder=5)
plt.scatter([0], [0], color='k', zorder=5) # large gravitating mass

# plot the trajectory
plt.plot(x, y, color='green', label="step = {}".format(dt), zorder=10)
plt.scatter([x[0]], [y[0]], color='green', zorder=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Semi-implicit Euler method")
plt.show()
```

## Method 5: Verlet

In this method, we write everything in terms of position only and cut out any reference to velocity. Let us derive it in two ways. Let us take the equation for the semi-implicit Euler's method, and plug $\vec{v}_{n+1} = v_{n} + a_{n}\Delta t$, and then plug $v_{n} = (x_{n}-x_{n-1})/\Delta t$: 
\begin{align*}
    \vec{x}_{n+1} &= \vec{x}_{n} + \vec{v}_{n+1}\Delta t \\
    &= \vec{x}_{n} + (\vec{v}_{n}+\vec{a}_{n}\Delta t)\Delta t \\
    &= \vec{x}_{n} + \vec{v}_{n}\Delta t + \vec{a}_{n}\Delta t^{2} \\
    &= \vec{x}_{n} + (\vec{x}_{n}-\vec{x}_{n-1})\tfrac{1}{\Delta t}\Delta t + \vec{a}_{n}\Delta t^{2} \\
    &= 2\vec{x}_{n} - \vec{x}_{n-1} + \vec{a}_{n}\Delta t^{2}.
\end{align*}
This gives us a way to update the position based on the last two positions. 

For the second derivation, take the Taylor series of the position:
\begin{align*}
    \vec{x}(t+\Delta t) &= \vec{x}(t) + \vec{v}(t)\Delta t + \frac{1}{2}\vec{a}(t)\Delta t^{2} + \frac{1}{6}\vec{b}(t)\Delta t^{3} + \cdots, \\
    \vec{x}(t-\Delta t) &= \vec{x}(t) - \vec{v}(t)\Delta t + \frac{1}{2}\vec{a}(t)\Delta t^{2} - \frac{1}{6}\vec{b}(t)\Delta t^{3} + \cdots.
\end{align*}
Adding these two together gives
\begin{align*}
    \vec{x}(t+\Delta t) + \vec{x}(t-\Delta t) = 2\vec{x}(t) + \vec{a}(t)\Delta t^{2} + O(\Delta t^{4}).
\end{align*}
Now we can use the system equation $\ddot{\vec{x}}(t) = \vec{a}(\vec{x}(t))$ to obtain our method. 

We have 
\begin{gather}
    \boxed{ \vec{x}_{n+1} = -\vec{x}_{n-1} + 2\vec{x}_{n} + \vec{a}(\vec{x}_{n})\Delta t^{2} }
\end{gather}
where
\begin{gather}
    \boxed{ \vec{x}_{1} = \vec{x}_{0} + \vec{v}_{0}\Delta t }
\end{gather}
and the initial conditions are $\vec{x}_{0} = (5.24, 0)$ and $\vec{v}_{0} = (0, 0.49)$. 

As we obtain from the second derivation, the local error is $O(\Delta t^{4})$, so this method is very effective and simple. 

```python
# Verlet
# step and initial conditions
totaltime = 10
dt = 0.1
x, y = np.array([5.24]), np.array([0.00])
vx, vy = np.array([0.00]), np.array([0.49])

# update rule
def accel_law(x, y):
    b = 5/np.power(x**2 + y**2, 1.5)
    return -b*x, -b*y

def Verlet_update(x, y, dt):
    if len(x) < 2 and len(y) < 2:
        raise ValueError("position arrays have less than two points")
    else:
        ax, ay = accel_law(x[-1], y[-1])
        newx, newy = 2*x[-1] - x[-2] + ax*dt*dt, 2*y[-1] - y[-2] + ay*dt*dt
        return np.append(x, newx), np.append(y, newy)

# create tajectory; should take `totaltime` seconds
# initial update
x, y = np.append(x, x[0]+vx[0]*dt), np.append(y, y[0]+vy[0]*dt)
# usual update
for i in range(0, round(totaltime/dt)):
    x, y = Verlet_update(x, y, dt)

# create chart with ellipse
plt.grid(color='lightgray', linestyle='--', zorder=0)
plt.xlim(-6, 6), plt.ylim(-6, 6)
plt.xticks(np.arange(-6, 7, 1)), plt.yticks(np.arange(-6, 7, 1))
plt.xlabel("x"), plt.ylabel("y")
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

theta = np.linspace(0, 2*np.pi, 100)
plt.plot(3.00*np.cos(theta)+2.24, 2.00*np.sin(theta), color='red', zorder=5)
plt.scatter([0], [0], color='k', zorder=5) # large gravitating mass

# plot the trajectory
plt.plot(x, y, color='green', label="step = {}".format(dt), zorder=10)
plt.scatter([x[0]], [y[0]], color='green', zorder=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Verlet method")
plt.show()
```
